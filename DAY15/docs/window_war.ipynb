## Day 15: Time-Series & Trend Smoothing

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../data/daily_registrations.csv', parse_dates=['Date'], index_col='Date')

df['3Day_MA'] = df['Registrations'].rolling(window=3).mean()
df['30Day_MA'] = df['Registrations'].rolling(window=30).mean()

plt.figure(figsize=(14, 6))
plt.plot(df['Registrations'], label='Actual Daily Sign-ups', alpha=0.25, color='gray')
plt.plot(df['3Day_MA'], label='3-Day Moving Average', color='blue', linewidth=2)
plt.plot(df['30Day_MA'], label='30-Day Moving Average', color='green', linewidth=2)
plt.title('Window War: 3-Day vs 30-Day Moving Average')
plt.xlabel('Date')
plt.ylabel('Registrations')
plt.legend()
plt.tight_layout()
plt.savefig('../docs/window_war.png', dpi=150)
plt.show()

print(f"Last 3-Day MA  : {df['3Day_MA'].iloc[-1]:.2f}")
print(f"Last 30-Day MA : {df['30Day_MA'].iloc[-1]:.2f}")
```

### Reflection: Why a 30-Day Moving Average is More Useful for Planning a June Marketing Push

The 3-Day Moving Average reacts sharply to every spike and dip in the data — it is highly sensitive to short-term noise such as weekend surges or a single viral campaign day. While this reactivity is useful for spotting immediate anomalies, it makes it unreliable for strategic planning because it amplifies random fluctuations rather than revealing the true underlying momentum of the platform.

The 30-Day Moving Average, on the other hand, smooths out all weekly seasonality and random noise, leaving only the genuine long-term growth trend. When planning a large marketing push for the first week of June, a strategist needs to understand the baseline trajectory of the business over the past month — not just what happened in the last 3 days, which could be skewed by a holiday, a technical outage, or a one-off promotion.

In short: **the 30-Day MA answers "where is MeetMux truly heading?", while the 3-Day MA only answers "what happened this week?"** Sound budget allocation for a June campaign must be grounded in the stable, noise-free trend that a 30-day window provides.
