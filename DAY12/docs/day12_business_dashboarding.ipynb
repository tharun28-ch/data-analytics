# Day 12: Business Dashboarding
## The Stakeholder Pivot

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("../data/sales_cleaned.csv")
df_north = df[df["Region"] == "North"]

fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle("Executive Business Performance Dashboard - North Region", fontsize=20)

sns.lineplot(ax=axes[0, 0], x="Date", y="Sales", data=df_north)
axes[0, 0].set_title("Sales Trend Over Time (North)")

sns.barplot(ax=axes[0, 1], x="Region", y="Sales", data=df_north, estimator=sum)
axes[0, 1].set_title("Revenue by Region (North)")

df_north["Category"].value_counts().plot.pie(ax=axes[1, 0], autopct="%1.1f%%")
axes[1, 0].set_title("Sales Distribution by Category (North)")

sns.scatterplot(ax=axes[1, 1], x="Price", y="Sales", data=df_north)
axes[1, 1].set_title("Price vs. Sales Volume (North)")

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

### Reflection
**Why is a 4-chart dashboard more effective for a manager than sending them 4 separate image files via email?**

A 4-chart dashboard consolidates all critical metrics into a single, cohesive view, allowing managers to instantly spot relationships across different dimensions. This unified layout minimizes cognitive load, avoids context switching, and clearly guides the narrative—from high-level trends to granular reasons—making data-driven executive decisions significantly faster.
