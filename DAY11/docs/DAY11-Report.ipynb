# Day 11: Correlation & Feature Relationships

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df = pd.read_csv("../data/sales_cleaned.csv")
numeric_df = df.select_dtypes(include=['float64', 'int64'])
corr_matrix = numeric_df.corr()

plt.figure(figsize=(10, 8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.show()
```

```python
sns.scatterplot(x='Discount', y='Sales', data=df)
plt.show()
```

```python
sns.regplot(x='Discount', y='Sales', data=df)
plt.show()
```

If 'Discount' and 'Sales' have a correlation of 0.85, it does not mean giving a discount DEFINITELY caused the sales to rise. Correlation does not imply causation. There might be hidden variables, or the relationship although correlated, could be driven by another factor like seasonality or promotions pulling generally more people to the store.
