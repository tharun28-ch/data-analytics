# Day 6: Advanced Visualizations & Dashboards

In this section, we'll explore advanced visualization techniques and create color-coded insights.

```python
# The Color Insight: Scatter Plot of productID vs Sales
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Ensure data is loaded
df = pd.read_csv("../data/sales.csv")

# Create scatter plot: productID vs Sales, colored by product
plt.figure(figsize=(12, 6))
sns.scatterplot(data=df, x='productID', y='Sales', hue='productID', s=150, palette='Set2', alpha=0.7)
plt.title("The Color Insight: Sales Distribution by Product", fontsize=14, fontweight='bold')
plt.xlabel("Product ID", fontsize=12)
plt.ylabel("Sales Amount ($)", fontsize=12)
plt.legend(title="Product ID", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

print("✓ Scatter Plot Created: Each product has its own color!")
```

## Reflection: How Box Plots Help Identify Data Errors

### Why This Matters for Our Data

Looking at our Scatter Plot, we can see **Product 3 (light blue)** has sales ranging from about **300 to 900** - that's a huge variation!

### The Problem with Bar Charts
A simple Bar Chart would only show:
- Product 1 average = ~650
- Product 2 average = ~350  
- Product 3 average = ~600
- Product 4 average = ~450
- Product 5 average = ~200

✗ **This hides the truth!** It just shows one number per product.

### The Power of Box Plots
A Box Plot reveals:
- **The box** = where 50% of data lives (typical range)
- **The line inside** = the median (middle value)
- **The whiskers** = the normal range
- **The dots** = OUTLIERS (extreme/unusual values) ⚠️

### Real Example from Our Data

Looking at **Product 3**:
- **Bar Chart says:** "Average sales = 600" ✓ Looks okay
- **Box Plot shows:** 
  - Most days: 300-500 (the box)
  - Some days: 800-900 (outliers as dots)
  - **This is INCONSISTENT!**

### Why Inconsistency is a Red Flag 🚩

**Products with consistent sales** (like Product 2):
- Same sales most days = predictable = reliable ✓

**Products with inconsistent sales** (like Product 3):
- Huge variation day to day = unpredictable = needs investigation
- Questions: Why the spike on certain days?
- Is this a supply issue? Demand issue?

### The Box Plot Advantage
**A Box Plot catches what Bar Charts miss:**
- Bar Chart: "Everything looks normal"
- Box Plot: "Wait... Product 3 has huge outliers! Something's going on!"

### Our Data Insight
From the Box Plot in Day 5, Product 3 is:
- ✅ Highest total revenue
- ❌ Most inconsistent day-to-day
- 🚩 Needs investigation: Why the extreme variation?

**Smart analysts use Box Plots to spot problems BEFORE they become bigger issues.**
